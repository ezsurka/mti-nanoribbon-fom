```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def kappa_u(m0,Mz):
    return -m0-Mz > 0
def kappa_l(m0,Mz):
    return -m0+Mz > 0
```

```python
m0 = np.linspace(-.1,.1,301)
Mz = np.linspace(-.1,.1,301)
```

```python
Mz_, m0_ = np.meshgrid(Mz,m0)
ku = kappa_u(m0_,Mz_)
kl = kappa_l(m0_,Mz_)
```

```python
num_edge_states = ku.astype(int) + kl.astype(int)
```

```python
plt.contourf(Mz_,m0_,num_edge_states)
plt.xlabel('$M_z$')
plt.ylabel('$m_0$')
plt.colorbar()
plt.savefig('num_edge_states.png')
```
